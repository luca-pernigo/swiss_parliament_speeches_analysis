```python
def df2docbin(df, nlp):
    docbin = DocBin()
    for index, row in df.iterrows():
        doc = nlp.make_doc(row['Name'])
        doc.cats = {'LABEL': row.label, 'NOT_LABEL': not row.label}
        docbin.add(doc)
    return docbin
```

```python
import spacy
from spacy.tokens import DocBin
import pandas as pd
import random
from sklearn.model_selection import train_test_split

train, test = train_test_split(names, shuffle=True, train_size=0.8, stratify=names.label)
```

```python
nlp = spacy.blank("en")

docbin_train = df2docbin(train, nlp)
docbin_train.to_disk('training.spacy')

docbin_test = df2docbin(test, nlp)
docbin_test.to_disk('test.spacy')
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

fig, axs = plt.subplots(math.ceil(7 / 2), 2, sharex=True, sharey=True, figsize=(25, 15))

for ax in axs.reshape(-1):
    ax.grid(False)
    ax.axis("off")
    ax.set_xticks([])
    ax.set_yticks([])

for i, year in enumerate(years):
    ax = axs.reshape(-1)[i]
    ax.set_title("Speeches/Pop " + f"{int(i) + 46}th" + " Legislature", fontsize=18)
    ch_year = ch[
        (ch["Year"] == f"{year}")
        | (ch["Year"] == f"{int(year) + 1}")
        | (ch["Year"] == f"{int(year) + 2}")
        | (ch["Year"] == f"{int(year) + 3}")
        | (ch["Year"] == f"{int(year) + 4}")
    ]

    ch_grouped = ch_year.groupby(["iso_3166_2"]).agg({"Count": "sum"}).reset_index()

    ch_year1 = ch[(ch["Year"] == f"{year}")]
    ch_year1 = ch_year1.sort_values("iso_3166_2")
    
    if year==1999 or year==2004 or year==2008:
        ch_year1["Count"] = (ch_grouped["Count"].values)/(census.Census2000.values)
        im = ch_year1.plot(column="Count", edgecolor="black", figsize=(20, 10), cmap="RdPu", ax=ax)
        
        
    else:
        ch_year1["Count"] = (ch_grouped["Count"].values)/(census.Census2010.values)
        im = ch_year1.plot(column="Count", edgecolor="black", figsize=(20, 10), cmap="RdPu", ax=ax)


norm = plt.Normalize(ch_year1["Count"].min(), ch_year1["Count"].max())
cmap = cm.RdPu
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])  

# Add color bar
cbar = fig.colorbar(sm, ax=axs.ravel().tolist(), orientation="horizontal", pad=-0.3)
cbar.ax.set_ylabel("Speech/Pop")  

plt.show()
```

```python
df_de=swissparl[swissparl["s_language_cld2"]=="de"][['s_transcript', 's_speaker_name',
       's_speaker_council', 's_speaker_canton', 's_speaker_faction',
       's_speaker_president', 'b_number', 'b_title_german', 'b_title_french',
      'd_date', 'd_time']]
```

```python
#        _   _             _   _             
#       | | | |           | | (_)            
#   __ _| |_| |_ ___ _ __ | |_ _  ___  _ __  
#  / _` | __| __/ _ \ '_ \| __| |/ _ \| '_ \ 
# | (_| | |_| ||  __/ | | | |_| | (_) | | | |
#  \__,_|\__|\__\___|_| |_|\__|_|\___/|_| |_|
                                           
# The next seven lines take a bit ~1.30 hour
```

```python
removal= ['ADV','PRON','CCONJ','PUNCT','PART','DET','ADP','SPACE', 'NUM', 'SYM']

tokens = []
for text in nlp_de.pipe(df_de.s_transcript):
    tok = [token.lemma_.lower() for token in text 
           if token.pos_ not in removal and not token.is_stop and token.is_alpha]
    tokens.append(tok)
```

```python
df_de['tokens'] = tokens
df_de
```

```python
from gensim.corpora.dictionary import Dictionary

dictionary = Dictionary(df_de['tokens'])
```

```python
#filter out low-frequency and high-frequency tokens, also limit the vocabulary to a max of 1000 words
dictionary.filter_extremes(no_below=10, no_above=0.4, keep_n=1000)
```

```python
corpus = [dictionary.doc2bow(doc) for doc in df_de['tokens']]
```

```python
from gensim.models import LdaMulticore

lda = LdaMulticore(corpus=corpus, id2word=dictionary, iterations=50, num_topics=5, workers=4, passes=10)
```

```python
import pyLDAvis.gensim_models
pyLDAvis.enable_notebook()


lda_display = pyLDAvis.gensim_models.prepare(lda, corpus, dictionary)
pyLDAvis.display(lda_display)
```

```python
#model output
topic_perc=lda.print_topics(-1)
```

```python
df_de['topic'] = [sorted(lda[corpus][text])[0][0] for text in range(len(df_de['s_transcript']))]
```
